```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import random as random
```

```python
wave_dat,ref_dat = np.loadtxt('ReflectanceData_Nb10V.csv', delimiter=',', skiprows=1, usecols= (0,1), unpack=True)
```

```python
plt.errorbar(wave_dat, ref_dat, fmt='-')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance')
plt.title("Reflectance vs Wavelength: Niobium 10 V")
plt.show() #using default size figure
```

```python
for i in np.arange(len(ref_dat)):
    randNum = random.randrange(-1,1,1)*0.005
    ref_dat[i]+=randNum

for i in np.arange(5):
    rand2 = random.randrange(0,700,1)
    for num in np.arange(rand2,rand2+100,3):
        rand3 = random.randrange(-5,5,1)*0.01
        ref_dat[num]+=rand3
```

```python
plt.errorbar(wave_dat, ref_dat, fmt='-')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance')
plt.title("Reflectance vs Wavelength: Niobium 10 V")
plt.show() #using default size figure
```

```python
volt = np.array([10,15,20,25,30])
arr1 = np.array([22,33,44,55,66])
arr2 = np.array([33.65,48.3,54.27,63.03,82.81])
```

```python
plt.errorbar(volt, arr1, fmt='o')
plt.errorbar(volt,arr2, fmt = 'o')
plt.errorbar(volt,arr1, fmt = '-')

a,b = np.polyfit(volt,arr2,1)
plt.plot(volt,a*volt+b)
plt.text(0,0,'y = ' + '{:.2f}'.format(b) + ' + {:.2f}'.format(a) + 'x', size=14)

plt.xlabel('Voltage (V)')
plt.ylabel('Thickness (nm)')
plt.title("Thickness v Voltage")
plt.legend(["Calculated", "Expected"])
plt.show() #using default size figure
```
